```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, RocCurveDisplay, auc
import tensorflow as tf
```

```python
df = pd.read_csv('../Data/balanced_sclaer_dataset_diabetes.csv')
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.isnull().sum().sum()
```

```python
y = df["Diabetes_012"]
X = df.drop(["Diabetes_012"], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3)
```

```python
y_train.to_numpy().shape
```

```python
print(X.shape[1])
```

```python
model_classification_1 = tf.keras.Sequential(
    [
        tf.keras.layers.Dense(64, activation="relu", input_shape=(X.shape[1],)),
        tf.keras.layers.Dense(128, activation="relu"),
        tf.keras.layers.Dropout(0.05),
        tf.keras.layers.Dense(64, activation="relu"),
        tf.keras.layers.Dense(32, activation="relu"),
        tf.keras.layers.Dense(16, activation="relu"),
        # используем 2 нейрон и softmax
        tf.keras.layers.Dense(3, activation="softmax"),
    ]
)
```

```python
model_classification_1.summary()
```

```python
model_classification_1.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.005), loss="sparse_categorical_crossentropy")
```

```python
model_classification_1.fit(X_train, y_train, epochs=25, verbose=None) #1m 4.5s
```

```python
model_classification_1.save("../models/keras_mode_file.h5")
```

```python
y_pred = np.around(model_classification_1.predict(X_test, verbose=None))
```

```python
y_pred = pd.DataFrame(data= y_pred)
```

```python
y_pred
```

```python
y_pred_arg = [np.argmax(pred) for pred in model_classification_1.predict(X_test, verbose=None)]
```

```python
y_pred_arg
```

```python
print(classification_report(y_test, y_pred_arg))
print(confusion_matrix(y_test, y_pred_arg))
```
