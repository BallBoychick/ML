```python
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
lematizer = WordNetLemmatizer()

def prepro_text(doc):
    lower_doc = []
    for i in doc:
        lower_doc.append(i.lower())

    punc_doc = []
    for i in lower_doc:
        punc_doc.append(''.join(c for c in i if c not in string.punctuation))
    
    s_w = set(stopwords.words('english'))
    filter_doc = []
    for text in punc_doc:
        words = text.split()
        filter_words = [word for word in words if word.lower() not in s_w]
        filter_text = ' '.join(filter_words)
        filter_doc.append(filter_text)

    lemm_doc = []
    for sentce in  filter_doc:
        words = word_tokenize(sentce)
        lemm_words = [lematizer.lemmatize(word) for word in words]
        lemm_sentce = ' '.join(lemm_words)
        lemm_doc.append(lemm_sentce)

    return lemm_doc
```

```python
prepro_messages = prepro_text(X)
```

```python
##softclust
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer
```

```python
tf_vectorizer = CountVectorizer(max_df=0.95, min_df=2, max_features=1000, stop_words='english')

tf = tf_vectorizer.fit_transform(X)

lda = LatentDirichletAllocation(n_components=10, max_iter=20, learning_method='online', learning_offset=50., random_state=0).fit(tf)
```

```python
tf_vectorizer.get_feature_names_out()
```

```python
def top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic #%d:" % topic_idx)
        print(" ".join([feature_names[i]
        for i in topic.argsort()[: n_top_words -1: 1]]))

top_words(lda, tf_vectorizer.get_feature_names_out(), 10)
```
